```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
with open('recipes.txt', 'r') as f:
    recipes = eval(f.read())
```

```python
item = recipes[0]
```

```python
item
```

```python
df = pd.DataFrame()
```

```python
for item in recipes:
    recipe_series = pd.Series(
        [item['title'], 
         item['information']['recipe-cuisine'], 
         item['information']['recipe-difficulty'], 
         item['image_url'], 
         ";".join([a['name'] for a in item['ingrs'] if 'name' in a]), 
         "\n".join(item['steps'])])
    
    df = df.append(recipe_series, ignore_index=True)
```

```python
columns = ['name', 'cuisine', 'difficulty', 'image_url', 'ingredients', 'steps']
df.columns = columns
```

```python
df.head()
```

```python
df.columns
```

```python
df.cuisine = df.cuisine.str.split(',').str[0]
```

```python
values = df.cuisine.value_counts()

plt.figure(figsize=(10, 7))
plt.ylim(0, 400)
sns.barplot(x=values.index, y=values.values)

plt.xticks(rotation=90)
plt.show()
```

```python
from paralleling.custom_func import parse
from paralleling.parallel_procesiing import ParallelPreprocessing
```

```python
p = ParallelPreprocessing(
    data=df,
    make_copy=False
)
```

```python
import re
import string
```

```python
def remove_punctuations(text):
    return re.sub(f'[{"".join(string.punctuation)}]', ' ', text)
```

### Шаги приготовления

```python
p.run_series_function('steps', remove_punctuations, paralleling=False, inplace=True)
```

```python
normal_steps = p.run_series_function('steps', parse)
```

```python
from nltk.corpus import stopwords
```

```python
russian_stopwords = stopwords.words('russian')
```

```python
def delete_stopwords_and_digits(texts):
    return [" ".join([word for word in text.split() if not word in russian_stopwords and not re.findall('\d', text)]) for text in texts]
```

```python
normal_steps_without_stopwords = delete_stopwords_and_digits(normal_steps)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer(min_df=5, max_df=200, max_features=200)
steps_vectors = vectorizer.fit_transform(normal_steps_without_stopwords).toarray()
```

### Ингридиенты

```python
p.data.ingredients = p.data.ingredients.str.replace(';', ' ')
```

```python
%%time
ingr_normal = p.run_series_function('ingredients', parse)
```

```python
ingr_normal_without_stopwords = delete_stopwords_and_digits(ingr_normal)
```

```python
vectorizer_ingr = TfidfVectorizer(max_features=30, ngram_range=(1, 2))
ingr_vector = vectorizer_ingr.fit_transform(ingr_normal_without_stopwords).toarray()
```

### Название

```python
name_normal = p.run_series_function('name', parse)
```

```python
name_normal_without_stopwords = delete_stopwords_and_digits(name_normal)
```

```python
vectorizer_name = TfidfVectorizer(min_df=10, max_df=300, max_features=30)
name_vector = vectorizer_name.fit_transform(name_normal_without_stopwords).toarray()
```

### Обучение модели

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
def get_features_by_categ_columns(categ_columns):
    cat_encoder = OneHotEncoder()
    X_cat = cat_encoder.fit_transform(df[categ_columns]).toarray()
    
    features = np.hstack((X_cat, steps_vectors, ingr_vector, name_vector))
    
    return features
```

```python
categ_columns = ['cuisine', 'difficulty']
```

```python
features = get_features_by_categ_columns(categ_columns)
```

```python
from sklearn.manifold import TSNE

tsne = TSNE(n_components=2)
transformed_features = tsne.fit_transform(features)
```

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
transformed_features_scaled = StandardScaler().fit_transform(transformed_features)
```

```python
dbscan = DBSCAN(eps=0.15)
dbscan.fit(transformed_features_scaled)
```

```python
plt.figure(figsize=(7, 5))

sns.scatterplot(
    x=transformed_features[:, 0],
    y=transformed_features[:, 1],
    hue=df['cuisine'],
    palette="Paired",
    legend=False
)
```

```python
plt.figure(figsize=(7, 5))

sns.scatterplot(
    x=transformed_features[:, 0],
    y=transformed_features[:, 1],
    hue=df['difficulty'],
    palette="Paired",
)
```

```python
plt.figure(figsize=(7, 5))

g = sns.scatterplot(
    x=transformed_features[:, 0],
    y=transformed_features[:, 1],
    hue=dbscan.labels_,
    palette="Paired",
)


g.legend(loc='center left', bbox_to_anchor=(1.25, 0.5), ncol=1)

plt.show()
```

```python
pd.DataFrame(features).to_csv('recipes_features_vkusno.csv', sep='\t', index=False)
```

## Что можно сделать с данными:

- Определение сложности по инструкции приготовления

```python
from sklearn.model_selection import train_test_split
```

```python
features = get_features_by_categ_columns(['cuisine'])
target = df['difficulty']
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.25)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(max_depth=10)
rfc.fit(X_train, y_train)
```

```python
rfc.score(X_test, y_test)
```
